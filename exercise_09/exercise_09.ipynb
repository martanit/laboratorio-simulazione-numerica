```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
x, y = np.loadtxt("TSP/circle_distance.dat", usecols=(0,1), delimiter=' ', unpack='True')
plt.title("L for best path")
plt.xlabel("Generation")
plt.ylabel("L")
plt.plot(x, y)
plt.grid()

plt.subplot(1,3,2)
x, y = np.loadtxt("TSP/circle_avg_distance.dat", usecols=(0,1), delimiter=' ', unpack='True')
plt.title("L averaged on best half population")
plt.xlabel("Generation")
plt.ylabel("L")
plt.plot(x, y)
plt.grid()

plt.subplot(1,3,3)
x, y = np.loadtxt("TSP/circle_path.dat", usecols=(0,1), delimiter=' ', unpack='True')
plt.title("Path")
plt.plot(x, y, '-o')

plt.tight_layout()
plt.show()
```
