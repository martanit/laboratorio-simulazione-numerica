### Exercise 03.1
- Compute at time $t=0$ **via Monte Carlo** the European call-option price, $C[S(0),0]$, and put-option price, $P[S(0),0]$:

1. <span style="color:red"> By sampling **directly** the final asset price $S(T)$</span> for a $GBM(r,\sigma^2)$

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats

# To read data from file
x, y, err_prog = np.loadtxt("data/output_es03.1.call.dat", \
                               usecols=(0,1,2), delimiter=' ', unpack='true')
plt.figure(figsize=(14,7))
plt.errorbar(x,y,yerr=err_prog)
# Plot horizontal line at y, from xmin to xmax
plt.hlines(14.975790778311286, 0, 100, colors='r', linestyles='dashed')
plt.title('European call-option')
plt.xlabel('#blocks')
plt.ylabel('price')
plt.grid(True)
plt.show()
```

```python
# To read data from file
x, y, err_prog = np.loadtxt("data/output_es03.1.put.dat", \
                               usecols=(0,1,2), delimiter=' ', unpack='true')
plt.figure(figsize=(14,7))
plt.errorbar(x,y,yerr=err_prog)
# Plot horizontal line at y, from xmin to xmax
plt.hlines(5.4595325819072364, 0, 100, colors='r', linestyles='dashed')
plt.title('European put-option')
plt.xlabel('#blocks')
plt.ylabel('price')
plt.grid(True)
plt.show()
```
