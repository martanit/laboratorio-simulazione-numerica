### Exercise 10.2

```python
plt.figure(figsize=(8,5))

x, y = np.loadtxt("es_10.2/circle_distance.dat", usecols=(0,1), delimiter=' ', unpack='True')
plt.plot(x, y, label='process 0')
x, y = np.loadtxt("es_10.2/circle_distance.dat", usecols=(0,2), delimiter=' ', unpack='True')
plt.plot(x, y, label='process 1')
x, y = np.loadtxt("es_10.2/circle_distance.dat", usecols=(0,3), delimiter=' ', unpack='True')
plt.plot(x, y, label='process 2')
x, y = np.loadtxt("es_10.2/circle_distance.dat", usecols=(0,4), delimiter=' ', unpack='True')
plt.plot(x, y, label='process 3')

plt.title("Circle: best path for different process")
plt.xlabel("Step")
plt.ylabel("L")
plt.grid()
plt.legend()

plt.show()
```

We obtain the same good results for each process.

```python

plt.figure(figsize=(8,5))

x, y = np.loadtxt("es_10.2/square_distance.dat", usecols=(0,1), delimiter=' ', unpack='True')
plt.plot(x, y, label='process 0')
x, y = np.loadtxt("es_10.2/square_distance.dat", usecols=(0,2), delimiter=' ', unpack='True')
plt.plot(x, y, label='process 1')
x, y = np.loadtxt("es_10.2/square_distance.dat", usecols=(0,3), delimiter=' ', unpack='True')
plt.plot(x, y, label='process 2')
x, y = np.loadtxt("es_10.2/square_distance.dat", usecols=(0,4), delimiter=' ', unpack='True')
plt.plot(x, y, label='process 3')

plt.title("Square: best path for different process")
plt.xlabel("Step")
plt.ylabel("L")
plt.grid()
plt.legend()

plt.show()
```
