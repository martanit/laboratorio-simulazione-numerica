```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
x, y = np.loadtxt("es_10.1/circle_distance.dat", usecols=(0,1), delimiter=' ', unpack='True')
plt.title("L for best path")
plt.xlabel("Step")
plt.ylabel("L")
plt.plot(x, y)
plt.grid()

plt.subplot(1,2,2)
x, y = np.loadtxt("es_10.1/circle_path.dat", usecols=(0,1), delimiter=' ', unpack='True')
plt.title("Path")
plt.plot(x, y, '-o')

plt.tight_layout()
plt.show()
```
