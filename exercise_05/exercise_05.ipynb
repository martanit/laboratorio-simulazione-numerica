### Exercise 05.1
- Use the Metropolis algorithm to sample $|\Psi_{1,0,0}(x,y,z)|^2$ and $|\Psi_{2,1,0}(x,y,z)|^2$ in cartesian coordinates using Bhor radius unit $a_0=5.29*10^{-11}$ and set an acceptance step in order to obtain a $\approx50\%$ ratio, starting to evaluate radius after equilibrium.

I sample the next position from uniform distribution on a sphere with radius of $a_0$ multiplied by an acceptance step to obtain an $\approx50\%$ acceptance ratio:

|    | acceptance step |
|----|-----------------|
| 1s |      $1.13$     |
| 2p |      $2.74$     |

Before starting to measure  $\left\langle r \right\rangle$, I equilibrate system with $10^4$ step. Then I perform simulation with $10^4$ step for each block, enough to obtain good results both for 1s and 2p radius. I set the origin as starting point for each block.

<span style="color:black"> 1. Esitmate of $\left\langle r \right\rangle_{\Psi_{1,0,0}}$ and its statistical uncertainties with a large number of throws $M=10^7$ as a function of the number of blocks, $N=100$.

```python
# To read data from file
x, y, err_prog = np.loadtxt("data/output_es05.1.1.dat", \
                               usecols=(0,1,2), delimiter=' ', unpack='true')
plt.figure(figsize=(14,7))
plt.errorbar(x,y,yerr=err_prog)
# Plot horizontal line at y, from xmin to xmax
plt.hlines(1.5, 0, 100, colors='r', linestyles='dashed')
plt.title('Ground state')
plt.xlabel('#blocks')
plt.ylabel('$<r>_{1s} [a_0]$')
plt.grid(True)
plt.show()
```

```python
x, y, z = np.loadtxt("data/output_es05.1.1.point.dat", \
                               usecols=(0,1,2), delimiter=' ', unpack='true')
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x, y, z, c=z, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 20)
plt.show()
```

If simulation start far from origin, for example set $x=y=z=10a_0$ the radius value converge to the right value for excited state, instead does not converge for the ground state.

```python
# To read data from file
plt.figure(figsize=(10,7))

plt.subplot(2,1,1)
x, y, err_prog = np.loadtxt("data/output_es05.1.1_far.dat", \
                               usecols=(0,1,2), delimiter=' ', unpack='true')
plt.hlines(1.5, 0, 100, colors='r', linestyles='dashed')
plt.errorbar(x,y,yerr=err_prog)
plt.grid(True)
plt.title('Ground state starting far')
plt.xlabel('#blocks')
plt.ylabel('$<r>_{1s} [a_0]$')

plt.show()             

plt.figure(figsize=(10,7))
plt.subplot(2,1,2)
x1, y1, err_prog1 = np.loadtxt("data/output_es05.1.2_far.dat", \
                               usecols=(0,1,2), delimiter=' ', unpack='true')
plt.hlines(5, 0, 100, colors='r', linestyles='dashed')
plt.errorbar(x1,y1,yerr=err_prog1)
plt.title('Excited state starting far')

plt.xlabel('#blocks')
plt.ylabel('$<r>_{2p} [a_0]$')
plt.grid(True)

plt.show()
```
